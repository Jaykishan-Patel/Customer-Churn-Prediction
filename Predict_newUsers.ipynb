```python
model = joblib.load("churn_random_forest_pipeline.pkl")
```

```python
def convert_tenure_to_bin(input_data, mx_tenure=72):
    """
    Convert tenure into tenure_group bins and remove tenure column.
    Logic: 1-12, 13-24, ..., up to mx_tenure
    """

    # Convert dict to DataFrame
    df = pd.DataFrame([input_data])

    # Create labels like 1-12, 13-24, ...
    labels = [f"{i}-{i+11}" for i in range(1, mx_tenure, 12)]

    # Create bins
    bins = list(range(1, mx_tenure + 12, 12))

    # Apply binning
    df["tenure_group"] = pd.cut(
        df["tenure"],
        bins=bins,
        right=False,
        labels=labels
    )

    # Drop original tenure
    df.drop("tenure", axis=1, inplace=True)

    return df
```

```python
input_data = {
    'gender': 'Female',
    'SeniorCitizen': 0,
    'Partner': 'Yes',
    'Dependents': 'No',
    'tenure':1,
    'PhoneService': 'No',
    'MultipleLines': 'No phone service',
    'InternetService': 'DSL',
    'OnlineSecurity': 'No',
    'OnlineBackup': 'Yes',
    'DeviceProtection': 'No',
    'TechSupport': 'No',
    'StreamingTV': 'No',
    'StreamingMovies': 'No',
    'Contract': 'Month-to-month',
    'PaperlessBilling': 'Yes',
    'PaymentMethod': 'Electronic check',
    'MonthlyCharges': 29.85,
    'TotalCharges': 29.85
}
```

```python
df=convert_tenure_to_bin(input_data)
```
